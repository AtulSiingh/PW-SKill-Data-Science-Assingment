
Q1-
 **Artificial Intelligence**:
Artificial Intelligence (AI) refers to the simulation of human intelligence processes by machines, particularly computer systems. These processes include learning, reasoning, problem-solving, perception, and language understanding. AI systems aim to perform tasks that typically require human intelligence.


**Machine Learning (ML)**:
Machine Learning is a subset of artificial intelligence that focuses on developing algorithms and models that allow computers to learn from data. Instead of being explicitly programmed to perform a task, a machine learning model learns patterns and relationships from data to make predictions or decisions. The primary goal is to enable computers to improve their performance on a specific task over time as they're exposed to more data.

**Deep Learning (DL)**:
Deep Learning is a subfield of machine learning that involves neural networks with multiple layers. It's inspired by the structure and function of the human brain's neural networks. Deep learning models, often referred to as deep neural networks, can automatically learn to represent and understand complex patterns in data, making them particularly effective in tasks like image and speech recognition. The term "deep" comes from the idea that these networks have many layers, allowing them to learn hierarchical representations.




Example: A self-driving car that can perceive its surroundings, make decisions based on that perception, and navigate through traffic without human intervention is an example of AI. It uses sensors, cameras, and algorithms to mimic human driving capabilities.

Q2- **Supervised Learning**:
Supervised learning is a machine learning paradigm where the algorithm is trained on a labeled dataset. Each input data point is paired with the correct corresponding output, allowing the algorithm to learn the relationship between inputs and outputs.

Examples:
1. **Email Spam Detection**: Given a labeled dataset of emails (spam or not), the algorithm learns to classify new emails as spam or not based on patterns in the text.
2. **Handwriting Recognition**: Training a model to recognize handwritten digits based on a dataset of labeled digits and their corresponding labels.
3. **Medical Diagnosis**: Using historical patient data to train a model to predict whether a patient has a specific medical condition based on various health attributes.

Q3- **Unsupervised Learning**:
Unsupervised learning involves training an algorithm on an unlabeled dataset, aiming to find patterns, structures, or relationships within the data without specific output labels.

Examples:
1. **Customer Segmentation**: Clustering customers into groups based on their purchasing behavior to understand different customer segments.
2. **Anomaly Detection**: Identifying unusual patterns or outliers in data, such as fraudulent transactions in a financial dataset.
3. **Topic Modeling**: Analyzing a collection of text documents to uncover underlying themes or topics without predefined categories.

Q4- **Difference between AI, ML, DL, and DS**:
- **AI (Artificial Intelligence)**: The broad concept of creating intelligent machines capable of human-like tasks.
- **ML (Machine Learning)**: A subset of AI that involves training algorithms to learn patterns from data.
- **DL (Deep Learning)**: A subfield of ML that uses neural networks with multiple layers to learn complex patterns.
- **DS (Data Science)**: The interdisciplinary field that involves extracting insights from data using techniques from statistics, ML, and domain knowledge.

Q5- **Differences between Supervised, Unsupervised, and Semi-Supervised Learning**:
- **Supervised Learning**: Uses labeled data for training and requires input-output pairs for learning.
- **Unsupervised Learning**: Utilizes unlabeled data to find patterns or structures.
- **Semi-Supervised Learning**: Combines labeled and unlabeled data for training, useful when labeled data is limited.

Q6- **Train, Test, and Validation Split**:
In machine learning, data is divided into:
- **Training Set**: Used to train the model.
- **Test Set**: Used to evaluate the model's performance on unseen data.
- **Validation Set**: Used for hyperparameter tuning and preventing overfitting.

Importance:
- Training Set: Teaches the model.
- Test Set: Measures its generalization to new data.
- Validation Set: Guides model improvement and prevents overfitting.

Q7- **Unsupervised Learning in Anomaly Detection**:
Unsupervised learning can identify anomalies by detecting data points that deviate significantly from the learned normal patterns within the dataset. Clustering and dimensionality reduction methods can help highlight anomalies.
